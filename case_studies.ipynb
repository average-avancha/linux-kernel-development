```python
import matplotlib.pyplot as plt
```

```python
# Function to accumulate page fault counts
def accumulate_page_faults(data):
    accumulated_counts = []
    total_faults_1 = 0
    total_faults_2 = 0

    for entry in data:
        jiffies, fault_1, fault_2, _ = map(int, entry.split())
        total_faults_1 += fault_1
        total_faults_2 += fault_2
        accumulated_counts.append((jiffies, total_faults_1 + total_faults_2))

    return accumulated_counts
```

```python
# Read data from file
with open('profile1.data', 'r') as file:
    lines = file.readlines()

# Extract and accumulate page fault counts
data = accumulate_page_faults(lines)

# Extract x and y values for plotting
jiffies, page_faults = zip(*data)

# Plot the graph
plt.plot(jiffies, page_faults)
plt.title('Accumulated Page Faults for Work Processes 1 and 2')
plt.xlabel('Time (Jiffies)')
plt.ylabel('Accumulated Page Faults')
plt.savefig('case_1_work_1_2.png')
plt.show()
```
